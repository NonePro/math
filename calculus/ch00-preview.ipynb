```python

%reload_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
from utils import *

# Define the number of sides and radius of the circle
num_sides = 8
radius = 5

points = distribute_points_on_circle(num_sides,radius,rotate=60)

fig, ax = plt.subplots()
ax.set_aspect(0.8)

points += [points[0]]
x_coords = [point[0] for point in points]
y_coords = [point[1] for point in points]

start = [points[0][0],points[0][1]] 
for i, point in enumerate(points):
    plt.plot([start[0],point[0]],[start[1],point[1]])
    if i > 1 and i < len(points)-2:
        mild = [(start[0]+point[0]+points[i+1][0])/3,(start[1]+point[1])/2] 
        plt.text(mild[0],mild[1],"A"+str(i-1))

plt.plot(x_coords, y_coords)
plt.axis("off")
```

```python

fig, axs = plt.subplots(2,4)
for i in range(2):
    for j in range(4):
        ax = axs[i][j]
        ax.set_aspect(1)

        edges = 4*i+j+3
        if edges == 8 or edges == 10:
            ax.text(0,0,"...")
            ax.axis("off")
            ax.plot(*circle_points(5),alpha=0)
            continue
        elif edges == 9:
            edges = 12

        ax.plot(*circle_points(5))
        polygon_points = distribute_points_on_circle(edges,5,90) 
        polygon_points += [polygon_points[0]] 
        x_coords = [point[0] for point in polygon_points]
        y_coords = [point[1] for point in polygon_points]
        ax.plot(x_coords,y_coords)
        ax.text(0,0,r"$A_{"+str(edges)+"}$")
        ax.axis("off")
    plt.tight_layout()
```
