# 为什么神经元几乎可以学习任何东西

```python
import numpy as np
import matplotlib.pyplot as plt 

def ReLU(x):
        return max(0,x) 

x = np.arange(-3, 3, 0.1)
y = x**3 + x**2 - x - 1 

plt.ylim(-3, 3)
plt.plot(x,y)

n1 = lambda x: -5*x -7.7 
n2 = lambda x : -1.2*x -1.3
n3 = lambda x: 1.2*x +1
n4 = lambda x: 1.2*x -0.2
n5 = lambda x: 2*x - 1.1
n6 = lambda x: 5*x-5

f = lambda x: -n1(x)-n2(x)-n3(x) + n4(x) + n5(x) + n6(x)
plt.plot(x,[f(x) for x in x])

plt.plot(x,[-n1(x) for x in x],'--',color='gray')
f1 = lambda x : - ReLU(n1(x))
plt.plot(x,[f1(x) for x in x],'--',color='red')
f2 = lambda x : - ReLU(n1(x))- ReLU(n2(x))
plt.plot(x,[f2(x) for x in x],'-',color="yellow")
f3 = lambda x : - ReLU(n1(x)) - ReLU(n2(x)) - ReLU(n3(x))
plt.plot(x,[f3(x) for x in x],'-.',color="green")
f6 = lambda x : - ReLU(n1(x)) - ReLU(n2(x)) - ReLU(n3(x)) + ReLU(n4(x)) + ReLU(n5(x)) + ReLU(n6(x))
plt.plot(x,[f6(x) for x in x],'-.',color="purple")

```
